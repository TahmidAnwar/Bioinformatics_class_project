```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis

# Define the two protein sequences
normal_seq = """
MAEPRQEFEVMEDHAGTYGLGDRKDQGGYTMHQDQEGDTDAGLKESPLQTPTEDGSEEPG
SETSDAKSTPTAEDVTAPLVDEGAPGKQAAAQPHTEIPEGTTAEEAGIGDTPSLEDEAAG
HVTQARMVSKSKDGTGSDDKKAKGADGKTKIATPRGAAPPGQKGQANATRIPAKTPPAPK
TPPSSGEPPKSGDRSGYSSPGSPGTPGSRSRTPSLPTPPTREPKKVAVVRTPPKSPSSAK
SRLQTAPVPMPDLKNVKSKIGSTENLKHQPGGGKVQIINKKLDLSNVQSKCGSKDNIKHV
PGGGSVQIVYKPVDLSKVTSKCGSLGNIHHKPGGGQVEVKSEKLDFKDRVQSKIGSLDNI
THVPGGGNKKIETHKLTFRENAKAKTDHGAEIVYKSPVVSGDTSPRHLSNVSSTGSIDMV
DSPQLATLADEVSASLAKQGL
"""

alzheimers_seq = """
MAEPRQEFEVMEDHAGTYGLGDRKDQGGYTMHQDQEGDTDAGLKESPLQTPTEDGSEEPGSETSDAKSTPTAEDVTAPLVDEGAPGKQAAAQPHTEIPEGTTAEEAGIGDTPSLEDEAAGHVTQARMVSKSKDGTGSDDKKAKGADGKTKIATPRGAAPPGQKGQANATRIPAKTPPAPKTPPSSGEPPKSGDRSGYSSPGSPGTPGSRSRTPSLPTPPTREPKKVAVVRTPPKSPSSAKSRLQTAPVPMPDLKNVKSKIGSTENLKHQPGGGKVQIIKKKLDLSNVQSKCGSKDNIKHVLGGGSVQIVYKPVDLSKVTSKCGSLGNIHHKPGGGQVEVKSEKLDFKDRVQSKIGSLDNITHVPGGGNKKIETHKLTFRENAKAKTDHGAEIVYKSPVVSGDTSPWHLSNVSSTGSIDMVDSPQLATLADEVSASLAKQGL
"""

# Function to calculate properties
def analyze_protein(seq, name):
    analysis = ProteinAnalysis(seq.replace("\n", "").strip())

    print(f"\n{name} Protein Analysis:")
    print("-" * 30)
    # Amino Acid Composition
    aa_composition = analysis.get_amino_acids_percent()
    print("Amino Acid Composition (%):")
    for aa, freq in aa_composition.items():
        print(f"  {aa}: {freq * 100:.2f}%")

    # Molecular Weight
    print(f"\nMolecular Weight: {analysis.molecular_weight():.2f} Da")

    # Isoelectric Point
    print(f"Isoelectric Point (pI): {analysis.isoelectric_point():.2f}")

    # Aromaticity
    print(f"Aromaticity: {analysis.aromaticity():.2f}")

    # Instability Index
    print(f"Instability Index: {analysis.instability_index():.2f} (Protein {'is stable' if analysis.instability_index() < 40 else 'is unstable'})")

    # Hydrophobicity (GRAVY)
    print(f"GRAVY (Hydrophobicity): {analysis.gravy():.2f}")

# Analyze both sequences
analyze_protein(normal_seq, "Normal Tau")
analyze_protein(alzheimers_seq, "Alzheimer's Tau")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from Bio.SeqUtils.ProtParam import ProteinAnalysis

# Define the two protein sequences
normal_seq = """
MAEPRQEFEVMEDHAGTYGLGDRKDQGGYTMHQDQEGDTDAGLKESPLQTPTEDGSEEPG
SETSDAKSTPTAEDVTAPLVDEGAPGKQAAAQPHTEIPEGTTAEEAGIGDTPSLEDEAAG
HVTQARMVSKSKDGTGSDDKKAKGADGKTKIATPRGAAPPGQKGQANATRIPAKTPPAPK
TPPSSGEPPKSGDRSGYSSPGSPGTPGSRSRTPSLPTPPTREPKKVAVVRTPPKSPSSAK
SRLQTAPVPMPDLKNVKSKIGSTENLKHQPGGGKVQIINKKLDLSNVQSKCGSKDNIKHV
PGGGSVQIVYKPVDLSKVTSKCGSLGNIHHKPGGGQVEVKSEKLDFKDRVQSKIGSLDNI
THVPGGGNKKIETHKLTFRENAKAKTDHGAEIVYKSPVVSGDTSPRHLSNVSSTGSIDMV
DSPQLATLADEVSASLAKQGL
"""
alzheimers_seq = """
MAEPRQEFEVMEDHAGTYGLGDRKDQGGYTMHQDQEGDTDAGLKESPLQTPTEDGSEEPGSETSDAKSTPTAEDVTAPLVDEGAPGKQAAAQPHTEIPEGTTAEEAGIGDTPSLEDEAAGHVTQARMVSKSKDGTGSDDKKAKGADGKTKIATPRGAAPPGQKGQANATRIPAKTPPAPKTPPSSGEPPKSGDRSGYSSPGSPGTPGSRSRTPSLPTPPTREPKKVAVVRTPPKSPSSAKSRLQTAPVPMPDLKNVKSKIGSTENLKHQPGGGKVQIIKKKLDLSNVQSKCGSKDNIKHVLGGGSVQIVYKPVDLSKVTSKCGSLGNIHHKPGGGQVEVKSEKLDFKDRVQSKIGSLDNITHVPGGGNKKIETHKLTFRENAKAKTDHGAEIVYKSPVVSGDTSPWHLSNVSSTGSIDMVDSPQLATLADEVSASLAKQGL
"""

# Function to calculate amino acid composition
def get_aa_composition(seq):
    analysis = ProteinAnalysis(seq.replace("\n", "").strip())
    return analysis.get_amino_acids_percent()

# Function to calculate physiochemical properties
def get_physiochemical_properties(seq):
    analysis = ProteinAnalysis(seq.replace("\n", "").strip())
    return {
        "Molecular Weight": analysis.molecular_weight(),
        "Isoelectric Point": analysis.isoelectric_point(),
        "Aromaticity": analysis.aromaticity(),
        "Instability Index": analysis.instability_index(),
        "GRAVY": analysis.gravy(),
    }

# Amino acid composition for both sequences
aa_composition_normal = get_aa_composition(normal_seq)
aa_composition_alzheimers = get_aa_composition(alzheimers_seq)

# Create data for bar plot
aa_composition_data = {
    "Amino Acid": list(aa_composition_normal.keys()),
    "Normal Tau": [v * 100 for v in aa_composition_normal.values()],
    "Alzheimer's Tau": [v * 100 for v in aa_composition_alzheimers.values()],
}

# Bar Plot for Amino Acid Composition
plt.figure(figsize=(12, 6))
x = range(len(aa_composition_data["Amino Acid"]))
plt.bar(x, aa_composition_data["Normal Tau"], width=0.4, label="Normal Tau", alpha=0.7)
plt.bar([i + 0.4 for i in x], aa_composition_data["Alzheimer's Tau"], width=0.4, label="Alzheimer's Tau", alpha=0.7)
plt.xticks([i + 0.2 for i in x], aa_composition_data["Amino Acid"])
plt.xlabel("Amino Acid")
plt.ylabel("Percentage Composition (%)")
plt.title("Amino Acid Composition Comparison")
plt.legend()
plt.show()

# Physiochemical properties for both sequences
physio_normal = get_physiochemical_properties(normal_seq)
physio_alzheimers = get_physiochemical_properties(alzheimers_seq)

# Create data for histogram
properties = list(physio_normal.keys())
values_normal = list(physio_normal.values())
values_alzheimers = list(physio_alzheimers.values())

# Bar Plot for Physiochemical Properties
plt.figure(figsize=(10, 5))
x = range(len(properties))
plt.bar(x, values_normal, width=0.4, label="Normal Tau", alpha=0.7)
plt.bar([i + 0.4 for i in x], values_alzheimers, width=0.4, label="Alzheimer's Tau", alpha=0.7)
plt.xticks([i + 0.2 for i in x], properties, rotation=45)
plt.ylabel("Value")
plt.title("Physiochemical Properties Comparison")
plt.legend()
plt.show()
```

```python
# Histogram for GRAVY and Instability Index
sns.histplot([physio_normal["GRAVY"], physio_alzheimers["GRAVY"]], bins=10, kde=True, label="GRAVY")
sns.histplot([physio_normal["Instability Index"], physio_alzheimers["Instability Index"]], bins=10, kde=True, label="Instability Index")
plt.legend(["GRAVY", "Instability Index"])
plt.title("Distribution of GRAVY and Instability Index")
plt.show()
```
